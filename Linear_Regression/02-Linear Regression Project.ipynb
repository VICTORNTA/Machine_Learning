## Training and Testing Data



```python
X = customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 90)
```

## Training The Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin =  LinearRegression()
```

```python
lin.fit(X,y)
```

```python
lin.coef_
```

## Predicting Test Data


```python
prediction = lin.predict(X_test)
```

```python
sns.scatterplot(y_test,lin.predict(X_test))
```

## Evaluating the Model


```python
from sklearn import metrics
```

```python
print ("MAE:%s" % metrics.mean_absolute_error(y_test,lin.predict(X_test)))
print ("RAE:%s" % np.sqrt(metrics.mean_squared_error(y_test,lin.predict(X_test))))
print ("MSE:%s" % metrics.mean_squared_error(y_test,lin.predict(X_test)))
```

```python
sns.distplot((y_test-prediction),bins = 50)
```
